```
import matplotlib.pyplot as plt
%matplotlib inline


import numpy as np
import seaborn as sns

x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```
# whitegrid 라는 스타일 지원

sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```
# Seaborn은 연습할 만한 데이터셋을 몇개 가지고 있음.
# tips라는 데이터 셋
# 요일별 점심, 저녁, 흡연 여부, 식사금액, 팁 정리한 테이블

tips = sns.load_dataset("tips")
tips.head()
```

```
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```
# 이렇게 boxplot을 그리는데 x축에는 요일로, y축에는 전체 금액
```

```
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data = tips, palette="Set3")
plt.show()
```

```
# hue라는 옵션을 이용해서 흡연여부로 구분
# 흡연자가 결제 금액의 범위가 크다는 것을 알 수 있음.
```

```
# darkgrid 스타일로 하고 lmplot을 그림
# 데이터를 scatter 처럼 그리고 직선으로 regression 한 그림도 같이 그려줌
# 유효범위도 ci로 잡아줌

sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

```
# lmplot도 hue 옵션을 가질수 있으며
# palette로 색상을 지정 가능

sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1", height=7)
plt.show()
```

```
# 다른 데이터 셋
# 연도 및 월별 항공기 승객수를 기록한 데이터

flights = sns.load_dataset("flights")
flights.head(5)
```

```
# pivot 기능으로 간편하게 월별, 연도별로 구분 할 수 있습니다.

flights = flights.pivot("month", "year","passengers")
flights.head()
```

```
# heatmap이라는 도구를 이용하면 이런 종류의 데이터는 그 경향을 설명하기 좋습니다.

plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

```
# 머신러닝에서 중요하게 다뤄지는 아이리스 꽃에 대한 데이터를 가지고 옵니다.
# 꽃잎, 꽃받침의 너비, 폭을 가지고 그 종을 구분 할 수 있는지를 알아보겠습니다.

sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

```
sns.pairplot(iris, hue="species")
plt.show()
```
